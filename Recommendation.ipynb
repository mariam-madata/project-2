Step 1: Importing the dependencies



```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

step 2: Data Collection and Pre-processing

```python
#Loading dthe data from csv file into pandas data frame
df=pd.read_csv('/content/movies.csv')
```

```python
#printing the first 5 rows of the dataframe
df.head()
```

```python
#Number of columns and rows in the dataframe
df.shape
```

```python
#selecting the relevant features for recommendation
Common_features=['genres','keywords','director']
print(Common_features)
```

```python
for i in Common_features:
  df[i]=df[i].fillna('')
```

```python
# #combining all the five commom feature
combined_features =df['genres']+' '+df['keywords']+' '+df['director']
print(combined_features)
```

```python
#converting the text data to  feature vectors
vectorizer = TfidfVectorizer()
feature_vectors=vectorizer.fit_transform(combined_features)
print(feature_vectors)
```

```python
#Getting the similarity scores using cosine similarity
similarity=cosine_similarity(feature_vectors)
print(similarity)
```

```python
print(similarity.shape)
```

```python
#Getting the movie name from the User
movie_name = input('Enter your favourite movie name')
```

```python
#creating alist with all the movie names given in the dataset
list_of_all_titles=df['title'].tolist()
print(list_of_all_titles)
```

```python
#Finding the close match for the movie name given by the user
find_close_match=difflib.get_close_matches(movie_name,list_of_all_titles)
print(find_close_match)
```

```python
close_match=find_close_match[0]
print(close_match)
```

```python
#find the index of the movie title
index_of_the_movie=df[df.title==close_match]['index'].values[0]
print(index_of_the_movie)
```

```python
similarity_score=list(enumerate (similarity [index_of_the_movie]))
print(similarity_score)
```

```python
sorted_similar_movies=sorted(similarity_score,key=lambda x:x[1], reverse = True)
print(sorted_similar_movies)
```

```python
# print the name of similar movies based on index

print('Movies suggested for you: \n')
i=1
for movie in sorted_similar_movies:
  index=movie[0]
  title_from_index=df[df.index==index]['title'].values[0]
  if(i<30):
    print (1,'.',title_from_index)
    i+=1
```
